<h3> Text Generator Class implemented from  scratch, this class generates text using different algorithms according to a markovian process : </h3>
<h6>
- Greedy approach ( sample = False)
<br>
- Sampling algorithm (sample = True)
<br>
- top_p(nuclues) (sample = True, top_p  = ]0 : 1 [ )
<br>
- top_k( sample = True, top_k > 0  )
<br>
</h6>


```python
class TextGenerator : 
    def __init__(self, corpus,sample=False,top_k = 0, top_p = 0.0) :
        self.text = corpus.lower()
        self.probs = {}
        self.prev_word = corpus.split(" ")[0]
        self.sample = sample
        self.top_k = top_k
        self.top_p = top_p
        
    def train(self) :
        for current_word in self.text.split(" ") :
            if current_word not in [".",",",""]:
                if self.prev_word not in self.probs.keys() :
                    self.probs[self.prev_word] = {}
                    self.probs[self.prev_word][current_word] = 1
                else : 
                    if current_word not in self.probs[self.prev_word].keys() : 
                        self.probs[self.prev_word][current_word] = 1
                    else : 
                        self.probs[self.prev_word][current_word] += 1

            self.prev_word = current_word

        for key in self.probs.keys() :
            sum_key = sum([ count_val for count_val in self.probs[key].values()]) 
            for next_key in self.probs[key] : 
                self.probs[key][next_key] /= sum_key
    
    def generate_text(self,input_text,num_generated_tokens) : 
        if self.sample == False : 

            return self.greedy_algo(input_text,num_generated_tokens)
        else :
            return self.sample_algo(input_text,num_generated_tokens)


    def greedy_algo(self, input_text, num_generated_tokens) :

        current = input_text.split(" ")[-1]
        generated_text = input_text


        for i in range(0,num_generated_tokens):
            if current not in self.probs:

                break

            propable_words = self.probs[current]

            max_prob = max(propable_words.values())
            candidates = [word for word, value in propable_words.items() if value == max_prob]

            probable_next_word = np.random.choice(candidates)


            generated_text += " " + probable_next_word

            current = probable_next_word

        return generated_text
    
    def sample_algo(self,input_text,num_generated_tokens) : 
        current = input_text.split(" ")[-1]
        generated_text = input_text

        for i in range(num_generated_tokens):
            if current not in self.probs:
                break




            susbequent_words = self.probs[current]



            if self.top_k >0 :

                sorted_words = sorted(susbequent_words.items(),key = lambda x : x[1],reverse=True)
                sorted_words = sorted_words[0:self.top_k]
                susbequent_words = dict(sorted_words)


            if self.top_p > 0:
                sorted_words = dict(sorted(susbequent_words.items(), key=lambda x: x[1], reverse=True))
                cumulative_probs = [sum(list(sorted_words.values())[0: i + 1]) for i in range(len(sorted_words))]

                for i in range(len(cumulative_probs)):
                    if self.top_p < cumulative_probs[i]:  # reaching the cumulative probability that covers top_p from the subsequent probs
                        susbequent_words = dict(list(sorted_words.items())[0: i + 1])
                        # redistributing the probable sample "subsequent_words"
                        total_prob_old_sample = sum(susbequent_words.values())

                        for key, val in susbequent_words.items():
                            susbequent_words[key] = val / total_prob_old_sample
                        
                        cumulative_probs = [sum(list(susbequent_words.values())[0: i + 1]) for i in range(len(susbequent_words))]

                        break

                        
            else  :
                 cumulative_probs = [ sum(list(susbequent_words.values())[0 : i+1]) for i in range(0, len(susbequent_words.keys())) ]
                


            # sampling randomly from [0 : 1] to select next_word in the list
            rand_int = np.random.random()

            chosen_index = 0

            for i in range(0,len(cumulative_probs)) : 
                if rand_int < cumulative_probs[i] :
                    chosen_index = i
                    break
            subsequent_word = list(susbequent_words.keys())[chosen_index]


            generated_text += " " + subsequent_word
            current =  subsequent_word


            
            


        return generated_text 
```

```python
textGen = TextGenerator(Text_corpus)
textGen.train()
input_text = 'young lady'
generated_text = textGen.generate_text(input_text,num_generated_tokens = 20)
print(generated_text)
```
